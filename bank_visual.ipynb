```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('http://13.234.66.67/summer19/datasets/bank.csv')
df.head(10)
```

```python
plt.xlabel('CreditScore')
plt.ylabel('Balance')
plt.bar(df['CreditScore'],df['Balance'])
plt.grid(color='green')
plt.show()
```

```python
plt.xlabel('Balance')
plt.ylabel('EstimatedSalary')
plt.scatter(df['Balance'],df['EstimatedSalary'],color='red')
plt.title('EstimatedSalary v/s Balance')
plt.show()
```

```python
plt.xlabel('HasCrCard')
plt.ylabel('IsActiveMember')
plt.grid()
plt.scatter(df['HasCrCard'],df['IsActiveMember'])
plt.title('IsActiveMember v/s HasCrCard' )
plt.show()
```
