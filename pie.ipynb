```python
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
x=[4,7,8,9,11,5]
y=[13,23,5,7,9,5]
player=['virat','dhoni','vijay','dhawan']
runs=[120,44,77,33]
exp=[0,0.1,0,0]
```

# ploting graph

```python
plt.xlabel('x-axes')
plt.ylabel('y-axis')
plt.scatter(x,y)
plt.grid(color='red')
plt.show()
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.pie(runs,labels=player,explode=exp,shadow=True,autopct='%1.1f%%')
plt.legend( loc='lower center')
plt.grid(color='green')
plt.show()
```
