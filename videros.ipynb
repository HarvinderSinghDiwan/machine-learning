```python
import cv2
```

```python
#starting camera
cap=cv2.VideoCapture(0)
```

```python
while cap.isOpened():
    status,frame=cap.read()
    #coverting image frame to gray-scale
    graying=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    print(frame.shape)
    #now we can draw all those patterns
    #line
    cv2.line(frame,(0,0),(200,300),(0,255,0),3)
    #rectangle
    cv2.rectangle(frame,(50,50),(250,300),(0,0,255),2)
    #circle
    cv2.circle(frame,(200,300),100,(255,0,0),2)
    #for text
    font = cv2.FONT_HERSHEY_SIMPLEX     #its font-style as:'FONT_HERSHEY_SIMPLEX  '
    cv2.putText(frame,'THESEUs',(10,500), font, 4,(255,255,255),2,cv2.LINE_AA) #here 'LINE_AA' is used for straiht-line text
    
    
    cv2.imshow('LIVE',frame)
    cv2.imshow('LIVEGRAY',graying)
    if cv2.waitKey(10) & 0xff ==ord('q'):    #here '0xff ==ord('q')' is used to take the keyboard-input for eliminating loop
        break
    
#cv2.destroyWindow('live')
cv2.destroyAllWindows()  #this is will destroy all Windows
    
```

```python
#to 
cap.release()
```

```python
cv2.imwrite('videoooos.jpg',frame)  #saves image as last appearing frame or image in video
```
