```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading data
df=pd.read_csv('salary.csv')
```

```python
#exp
exp=df.iloc[:,0:1].values
```

```python
#sal
sal=df.iloc[:,1:].values
```

```python
#call linear model just for cmaparison
from sklearn.linear_model import LinearRegression
regrl=LinearRegression()
```

```python
#to transfer data we can call poly transform method
from sklearn.preprocessing import PolynomialFeatures
```

```python
#experince transform
poly=PolynomialFeatures(degree=4)#its make dgree=4polynomial (quartic-one)
```

```python
newexp=poly.fit_transform(exp)  #fit exp
```

```python
#apply linear regression
trained=regrl.fit(exp,sal)
```

```python
#now creating linear model again
plt.xlabel("EXPERIENCE")
plt.ylabel("SALARy")
plt.scatter(exp,sal,label="actual salary")
plt.plot(exp,trained.predict(exp),label="linear model predicted salary",color='brown')
plt.legend()
plt.show()
```

```python
#apply newexp data in model
regrp=LinearRegression()
```

```python
#new exp data apply
trained2=regrp.fit(newexp,sal)
```

```python
#now creating model again
plt.xlabel("EXPERIENCE")
plt.ylabel("SALARy")
plt.scatter(exp,sal,label="actual salary")
plt.plot(exp,trained2.predict(newexp),label="linear model predicted salary by polynomial:newexp",color='brown')
plt.legend()
plt.grid(color='green')
plt.show()
```
